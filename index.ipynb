### Piecharts

```python
sorted_count = pokemon['generation_id'].value_counts()
labels = ['gen{}'.format(x) for x in sorted_count.index]
plt.pie(sorted_count, labels=labels, startangle = 90, counterclock = False);
plt.axis('square');
```
